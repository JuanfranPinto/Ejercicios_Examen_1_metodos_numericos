```python
def secant_method(f, x0, x1, tol=1e-6, max_iter=100):
    

    # Contador de llamadas a f
    f_calls = 0

    # Evaluar inicialmente una vez cada punto
    f_prev = f(x0); f_calls += 1
    f_curr = f(x1); f_calls += 1

    x_prev, x_curr = x0, x1
    iter_count = 0

    while abs(f_curr) > tol and iter_count < max_iter:

        # Evitar divisiÃ³n entre cero
        denom = f_curr - f_prev
        if denom == 0:
            raise ZeroDivisionError("Division by zero in the secant method.")

        # FÃ³rmula optimizada
        x_next = x_curr - f_curr * (x_curr - x_prev) / denom

        # Avanzar valores
        x_prev, x_curr = x_curr, x_next
        f_prev, f_curr = f_curr, f(x_next); f_calls += 1  # SOLO 1 nueva llamada

        iter_count += 1

    return x_curr, iter_count, f_calls
```
